```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with relevant columns
data = {
    'screen_time': np.random.randint(30, 300, 100),
    'gender': np.random.choice(['Male', 'Female', 'Other'], 100),
    'location': np.random.choice(['Urban', 'Rural', 'Suburban'], 100),
    'device_type': np.random.choice(['Mobile', 'Tablet', 'Desktop'], 100)
}
df = pd.DataFrame(data)

# Code for Day 1 visualization will go here
# Example: Plotting distribution of screen time by gender
import seaborn as sns
import matplotlib.pyplot as plt
# Assuming 'df' is your DataFrame with 'screen_time' and 'gender' columns
sns.histplot(data=df, x="screen_time", hue="gender", kde=True)
plt.title("Distribution of Screen Time by Gender")
plt.show()
```

**Day 2: Stacked barplots (device usage by location)**

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with relevant columns
data = {
    'screen_time': np.random.randint(30, 300, 100),
    'gender': np.random.choice(['Male', 'Female', 'Other'], 100),
    'location': np.random.choice(['Urban', 'Rural', 'Suburban'], 100),
    'device_type': np.random.choice(['Mobile', 'Tablet', 'Desktop'], 100)
}
df = pd.DataFrame(data)

# Code for Day 2 visualization will go here
# Example: Plotting stacked barplot of device usage by location
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Assuming 'df' is your DataFrame with 'location' and 'device_type' columns
device_location_counts = df.groupby(['location', 'device_type']).size().unstack(fill_value=0)
device_location_counts.plot(kind='bar', stacked=True)
plt.title("Device Usage by Location")
plt.xlabel("Location")
plt.ylabel("Number of Users")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Day 3: Swarmplots/stripplots (screen time spread)**

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with relevant columns
data = {
    'screen_time': np.random.randint(30, 300, 100),
    'gender': np.random.choice(['Male', 'Female', 'Other'], 100),
    'location': np.random.choice(['Urban', 'Rural', 'Suburban'], 100),
    'device_type': np.random.choice(['Mobile', 'Tablet', 'Desktop'], 100)
}
df = pd.DataFrame(data)

# Code for Day 3 visualization will go here
# Example: Plotting swarmplot or stripplot for screen time spread
import seaborn as sns
import matplotlib.pyplot as plt
# Assuming 'df' is your DataFrame with a 'screen_time' column
sns.swarmplot(data=df, y="screen_time") # or sns.stripplot(data=df, y="screen_time")
plt.title("Screen Time Spread")
plt.show()
```

**Day 4: Combine multiple plots into subplots**

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with relevant columns
data = {
    'screen_time': np.random.randint(30, 300, 100),
    'gender': np.random.choice(['Male', 'Female', 'Other'], 100),
    'location': np.random.choice(['Urban', 'Rural', 'Suburban'], 100),
    'device_type': np.random.choice(['Mobile', 'Tablet', 'Desktop'], 100)
}
df = pd.DataFrame(data)

# Code for Day 4 visualization will go here
# Example: Combining multiple plots into subplots
import matplotlib.pyplot as plt
import seaborn as sns
# Assuming 'df' is your DataFrame with 'screen_time' and 'gender' columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
sns.histplot(data=df, x="screen_time", hue="gender", kde=True, ax=axes[0])
axes[0].set_title("Distribution of Screen Time by Gender")
sns.swarmplot(data=df, y="screen_time", ax=axes[1])
axes[1].set_title("Screen Time Spread")
plt.tight_layout()
plt.show()
```
