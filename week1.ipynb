```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path='/content/drive/MyDrive/Indian_Kids_Screen_Time.csv'
df=pd.read_csv(file_path)

#Violin plot for Screen Time by Age
plt.figure(figsize=(10,5))
sns.violinplot(data=df, x='Age', y='Avg_Daily_Screen_Time_hr', palette="muted", hue='Age', legend=False)
plt.title('Daily Screen Time Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Avg Daily Screen Time (hrs)')
plt.show()
```

```python
# Boxplot: Screen Time by Age
plt.figure(figsize=(10, 6))
sns.boxplot(x="Age", y="Avg_Daily_Screen_Time_hr", data=df, palette="Set2", hue='Age', legend=False)
plt.title("Boxplot of Screen Time by Age", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Average Daily Screen Time (hr)")
plt.show()
```

```python
# Boxplot: Screen Time by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x="Gender", y="Avg_Daily_Screen_Time_hr", data=df, palette="pastel", hue='Gender', legend=False)
plt.title("Boxplot of Screen Time by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Average Daily Screen Time (hr)")
plt.show()
```

```python
sns.set(style="whitegrid")

# Histogram for Age
plt.figure(figsize=(7, 5))
sns.histplot(df['Age'], bins=15, kde=False, color='skyblue')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram for Average Daily Screen Time
plt.figure(figsize=(7, 5))
sns.histplot(df['Avg_Daily_Screen_Time_hr'], bins=15, kde=False, color='lightcoral')
plt.title("Histogram of Average Daily Screen Time (hr)")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Count")
plt.show()
```

```python
# KDE Plot for Age
plt.figure(figsize=(7, 5))
sns.kdeplot(df['Age'], fill=True, color='blue')
plt.title("KDE Plot of Age")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.kdeplot(df['Avg_Daily_Screen_Time_hr'], fill=True, color='red')
plt.title("KDE Plot of Average Daily Screen Time (hr)")
plt.xlabel("Screen Time (hours)")
plt.ylabel("Density")
plt.show()
```
