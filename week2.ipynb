```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv('/content/Indian_Kids_Screen_Time.csv')

# Display the first 5 rows of the DataFrame
display(df.head())

# Display the data types of each column
display(df.info())

# Display summary statistics of the DataFrame
display(df.describe())
```

## Day 1: histograms & kde plots

### Subtask:
Generate histograms and KDE plots for continuous variables like 'screen time' and 'age'.


**Reasoning**:
Generate histograms with KDE overlays for 'Age' and 'Avg_Daily_Screen_Time_hr' to visualize their distributions.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Age', kde=True)
plt.title('Distribution of Age')

plt.subplot(1, 2, 2)
sns.histplot(data=df, x='Avg_Daily_Screen_Time_hr', kde=True)
plt.title('Distribution of Average Daily Screen Time')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Gender')

plt.subplot(1, 3, 2)
sns.countplot(data=df, x='Primary_Device')
plt.title('Distribution of Primary Device')

plt.subplot(1, 3, 3)
sns.countplot(data=df, x='Urban_or_Rural')
plt.title('Distribution of Urban or Rural Location')

plt.tight_layout()
plt.show()
```

## Day 3: boxplots & violin plots

### Subtask:
Generate boxplots and violin plots to visualize the relationship between 'screen time' and categorical variables like 'gender' and 'age group'.


**Reasoning**:
Generate boxplots and violin plots to visualize the relationship between 'screen time' and categorical variables like 'gender' and 'age group' in two separate figures with two subplots each.



```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Gender', y='Avg_Daily_Screen_Time_hr')
plt.title('Screen Time vs. Gender (Boxplot)')

plt.subplot(1, 2, 2)
sns.violinplot(data=df, x='Gender', y='Avg_Daily_Screen_Time_hr')
plt.title('Screen Time vs. Gender (Violin Plot)')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Age', y='Avg_Daily_Screen_Time_hr')
plt.title('Screen Time vs. Age (Boxplot)')

plt.subplot(1, 2, 2)
sns.violinplot(data=df, x='Age', y='Avg_Daily_Screen_Time_hr')
plt.title('Screen Time vs. Age (Violin Plot)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Avg_Daily_Screen_Time_hr')
plt.title('Relationship between Age and Average Daily Screen Time')
plt.show()
```
